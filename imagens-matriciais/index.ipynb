```python
#Convert the image to a binary image
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

#for every pixel in the image, if the pixel value is greater than a threshold, set it to white, otherwise set it to black
def binarize(img, threshold):
    img_bin = np.zeros(img.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i, j] > threshold:
                img_bin[i, j] = 255
            else:
                img_bin[i, j] = 0
    return img_bin

img = Image.open('blade_gray.jpg')
img_np = np.array(img)
img_avg2 = np.average(img_np[:, :, :3], axis=2)#.astype(np.uint8)

img_bin = binarize(img_avg2, 100)
plt.imshow(img_bin, cmap='gray')
```
